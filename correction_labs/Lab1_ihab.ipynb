## Correction Lab 1
- Question 1

```python
%%bigquery
SELECT
  subject,
  author.name AS author_name,
  committer.date AS commit_date,
  message
FROM
  `bigquery-public-data.github_repos.sample_commits`
ORDER BY
  committer.date DESC
LIMIT 10;
```

- Question 2

```python
%%bigquery
SELECT
  author.name AS author_name,
  COUNT(*) AS total_commits
FROM
  `bigquery-public-data.github_repos.sample_commits`

GROUP BY
  author_name
HAVING author_name is not null
ORDER BY
  total_commits DESC
LIMIT 10;
```

- Question 3

```python
%%bigquery
SELECT
  FORMAT_DATE('%Y-%m', DATE(committer.date)) AS year_month,
  COUNT(*) AS commit_count
FROM
  `bigquery-public-data.github_repos.sample_commits`
GROUP BY
  year_month
ORDER BY
  year_month DESC;
```

- question 4

```python
%%bigquery
SELECT
  author.name AS author_name,
  committer.date AS commit_date,
  message
FROM
  `bigquery-public-data.github_repos.sample_commits`
WHERE
  REGEXP_CONTAINS(LOWER(message), r'\b(fix|bug)\b')
ORDER BY
  committer.date DESC
LIMIT 50;
```

- Question 5

```python
%%bigquery
SELECT
  author.name AS author_name,
  AVG(LENGTH(message)) as avg_message_length,
  COUNT(*) AS total_commits
FROM
  `bigquery-public-data.github_repos.sample_commits`
GROUP BY
  author_name
HAVING author_name is not null
ORDER BY
  avg_message_length DESC
LIMIT 20;
```

- Question 6

```python
%%bigquery
SELECT
  author.name AS author_name,
  committer.date AS commit_date,
  message,
  LENGTH(message) AS message_length
FROM
  `bigquery-public-data.github_repos.sample_commits`
WHERE
  LENGTH(message) < 15
ORDER BY
  message_length ASC
LIMIT 50;
```
